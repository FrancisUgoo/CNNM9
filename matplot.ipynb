```python
x = np.array([1,2,3])
y = np.array([0.9100,0.9500,0.9700])

print(len(x))
print(len(y))

font1 = {'family':'serif','color':'blue','size':12}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Showing Avg accuracy of each epoch", fontdict = font1)
plt.xlabel("Number of Epochs", fontdict = font2)
plt.ylabel("Accuracy", fontdict = font2)

plt.plot(x, y, 'go-')

plt.grid(color = 'black', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([0.3016,0.1472,0.0892])

print(len(x))
print(len(y))

font1 = {'family':'serif','color':'blue','size':12}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Showing Avg Loss of each epoch", fontdict = font1)
plt.xlabel("Number of Epochs", fontdict = font2)
plt.ylabel("Loss", fontdict = font2)

plt.plot(x, y, 'r*-')

plt.grid(color = 'black', linestyle = '--', linewidth = 0.5)

plt.show()
```
